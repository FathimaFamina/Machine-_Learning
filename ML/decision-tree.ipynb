```python
import seaborn as sns
df=sns.load_dataset('iris')
df.head()
```

```python
df.columns
```

```python
#independent feature and dependent features
X=df.iloc[:,:-1]
y=iris.target
```

This line selects all rows (:) and all columns except the last one (:-1) from the DataFrame df.
So, X will contain all the features (independent variables) from your dataset df, excluding the last column.

```python
X,y
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel=DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(13,8))
tree.plot_tree(treemodel,filled=True)
```

Splitting Criterion: x[2] <= 2.45
This indicates the condition or rule that the decision tree uses to divide the dataset at a particular node.
x[2] refers to the feature at index 3 in your dataset X. In machine learning models, features are typically indexed starting from 0.
<= 2.45 specifies the threshold value for the featurex[2]. The decision tree will split the data such that one subset will have values of x[2] less than or equal to 2.45, and the other subset will have values greater than 2.45.

Gini Impurity: gini = 0.666.Gini impurity is a measure of how often a randomly chosen element from the set would be incorrectly labeled if it was randomly labeled according to the distribution of labels in the subset.
It ranges from 0 to 1, where 0 indicates that all elements in a set are the same class, and 1 indicates that the elements are uniformly distributed across all classes.
In this case, gini = 0.666 means that the impurity (or uncertainty) at this node is 0.666 or 66.6%. This suggests that the samples at this node are not perfectly classified; there is a significant mixture of different classes.

Node 1:

```python
# Splitting criterion: x[3] <= 0.8
# Gini impurity at this node: gini = 0.666
# Number of samples at this node: samples = 100
# Class distribution of samples at this node: value = [31, 35, 34]
# This means there are 31 samples of class 0, 35 samples of class 1, and 34 samples of class 2.
```

Node 2 (Child node of Node 1):

```python
# Gini impurity: gini = 0.0
# Number of samples: samples = 31
# Class distribution: value = [31, 0, 0]
# All 31 samples at this node belong to class 0.
```

Node 3 (Child node of Node 1):

```python
# Splitting criterion: x[3] <= 1.75
# Gini impurity: gini = 0.5
# Number of samples: samples = 69
# Class distribution: value = [0, 35, 34]
# There are 35 samples of class 1 and 34 samples of class 2 at this node.
```

```python
#prediction using test data
y_pred=treemodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
